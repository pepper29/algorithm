```python
'''
ğŸ‘š ìë£Œ ì •ë¦¬: ì£¼ì–´ì§„ clothes ë°°ì—´ì„ ë°˜ë³µë¬¸ìœ¼ë¡œ ìˆœíšŒí•˜ë©°, ë”•ì…”ë„ˆë¦¬ì— 'ì¢…ë¥˜'ë¥¼ í‚¤ë¡œ, 'ê°œìˆ˜'ë¥¼ ê°’ìœ¼ë¡œ ì €ì¥í•©ë‹ˆë‹¤.

âœ–ï¸ ê³±ì…ˆ ê³„ì‚°: ë”•ì…”ë„ˆë¦¬ì— ì €ì¥ëœ ëª¨ë“  'ê°œìˆ˜' ê°’ì— 1ì„ ë”í•œ í›„, ì´ ê°’ë“¤ì„ ëª¨ë‘ ê³±í•˜ì—¬ ì´ ê²½ìš°ì˜ ìˆ˜(ì…ì§€ ì•Šì€ ê²½ìš° í¬í•¨)ë¥¼ ê³„ì‚°í•©ë‹ˆë‹¤.

â– ìµœì¢… ì²˜ë¦¬: ê³„ì‚°ëœ ì´ ê²½ìš°ì˜ ìˆ˜ì—ì„œ 1ì„ ë¹¼ì„œ ìµœì†Œ í•œ ë²Œ ì´ìƒ ì…ëŠ” ì¡°í•©ì˜ ê°œìˆ˜ë¥¼ êµ¬í•©ë‹ˆë‹¤.
'''

def solution(clothes):
    # Step.1 ë”•ì…”ë„ˆë¦¬ì— ë§ì¶°ì£¼ê¸°
    answer = 1
    d_c = {}
    for d in clothes:
        print('í™•ì¸í•  key,value:',d)
        if d[1] in d_c:
            d_c[d[1]]+=1
            print('ê¸°ì¡´ê±°ì— ì¶”ê°€í•˜ëŠ” key',d_c[d[1]])
            print('í˜„ì¬dict',d_c)
        else:
            d_c[d[1]]=1
            print('ìƒˆë¡­ê²Œ ì¶”ê°€í•˜ëŠ” key',d_c[d[1]])
            print('í˜„ì¬dict.key:',d_c)

    # Step.2 0ì¼ì¡°ê±´ìœ¼ë¡œ +1ì”© í•´ì£¼ê³  ëª¨ë“  ê²½ìš°ì˜ ìˆ˜ ê³±ì…ˆ
    print(d_c.values()) # ë”í•˜ê¸° ì „

    for com in d_c.values():
        answer = answer * (com+1)
    print('ëª¨ë“  ê²½ìš°ì˜ ìˆ˜:',answer)

    # Step.3 ëª¨ë“  ì¡°ê±´ì—ì„œ ì•„ë¬´ê²ƒë„ ì„ íƒë˜ì§€ ì•Šì„ ì¡°ê±´ -1
    answer = answer-1
    return answer
```
