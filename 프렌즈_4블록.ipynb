```python
# Step 1: 블록 검색
# Step 2: 블록 제거 및 점수 집계
# Step 3: 블록 떨어뜨리기
# m: 높이, n:너비
import numpy as np

def solution(m, n, board):
    # 행렬로 배열정의
    grid = np.array([list(row) for row in board])
    total_removed = 0 # 지워진 블록 카운트
    #print(grid) # 출력

    # 정의 1: 블록 검색
    def find_block(grid, m, n):
        # 지워지는 배열 정의.
        to_remove = set() # 중복 안되게 set()
        for i in range(m-1): # 높이
            for j in range(n-1): # 너비
                target = grid[i,j] # 현재위치 (i, j)
                # 현재 칸이 비어있지 않고, 주변 3칸과 모두 같은지 확인
                if target != '0' and (grid[i:i+2, j:j+2] == target).all():
                    # 기존 값 0으로 지우기
                    to_remove.update([ (i,j), (i+1,j), (i,j+1), (i+1,j+1) ])
        return to_remove
    #print(find_block(grid,m,n))

    # 정의 2: 블록 떨어뜨리기
    def drop_block(grid,m,n):
        # 각 열을 하나씩 체크
        for e in range(n):
            # 1. e열에서 0이 아닌 블록들만 추출
            # grid[:, e]로 특정 열을 가져올 수 있음
            blocks = [char for char in grid[:, e] if char != '0']

            # 2. 부족한 칸수만큼 앞에 0을 채워서 새로운 열 생성
            # 전체 높이(m)에서 추출한 블록 개수를 뺴면 빈칸의 개수
            new_col = ['0'] * (m - len(blocks)) + blocks

            # 3. 새로 만든 열을 원래 grid의 해당 열에 다시 대입
            grid[:, e] = new_col
        return grid

    # main 함수
    while True:
        # Step 1: 블록 검색
        to_removed = find_block(grid, m, n)

        # Step 2: 중단조건
        if not to_removed:
            break

        # Step 3: 블록 지우기 및 개수 누적
        total_removed += len(to_removed)
        for q, w in to_removed:
            grid[q, w] = '0'

        # Step 4: 블록 떨구기
        grid = drop_block(grid, m, n)

    return total_removed
```
